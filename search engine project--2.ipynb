# Preprocessing Subtitles

```python
import string
import re
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
wl = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

def process(subtitle):
    # Clean the subtitle content
    subtitle_clean = re.sub("[^a-zA-Z\?\!]", " ", subtitle)
    subtitle_clean = subtitle_clean.lower()
    
    # Tokenize the subtitle into sentences
    sentences = sent_tokenize(subtitle_clean)
    
    # Process each sentence
    processed_sentences = []
    for sentence in sentences:
        # Tokenize the sentence into words
        words = word_tokenize(sentence)
        
        # Lemmatize and remove stop words
        words = [wl.lemmatize(word, pos='v') for word in words if word not in stop_words]
        
        # Join the processed words into a sentence
        processed_sentence = " ".join(words)
        processed_sentences.append(processed_sentence)
    
    # Join the processed sentences into a single string
    processed_subtitle = " ".join(processed_sentences)
    return processed_subtitle

# Apply the process function to each subtitle in the DataFrame
subtitle_data['subtitle_processed']=subtitle_data['file_content'].apply(process)
```

# Chunking and Final Processing

```python
def custom_chunker(chunk_size, unique_id, content, name, chunk_buffer):
    # Split content into tokens
    tokens = [token for token in content.split()]
    # Get the length of the tokens
    num_tokens = len(tokens)
    
    # Initialize index and count
    start_index, chunk_count = 0, 0
    
    while True:
        if start_index == 0:
            chunk_start, chunk_end = start_index,chunk_buffer
            if chunk_end > num_tokens:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:]))
                names.append(name)  # Append the name here
                break
            else:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:chunk_end]))
                names.append(name)  # Append the name here
                start_index = chunk_end
        else:
            chunk_start, chunk_end = start_index - chunk_size, start_index + chunk_buffer - chunk_size
            if chunk_end > num_tokens:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:]))
                names.append(name)  # Append the name here
                break
            else:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:chunk_end]))
                names.append(name)  # Append the name here
                start_index = chunk_end
        chunk_count += 1

new_id = []
new_content = []
names = []

for i in range(len(df_subtitle)):
    custom_chunker(25, i, df_subtitle['subtitle_processed'][i], df_subtitle['name'][i], 250)

new_df = pd.DataFrame({'id': new_id, 'subtitle_content': new_content, 'name': names})
print(new_df)
```
