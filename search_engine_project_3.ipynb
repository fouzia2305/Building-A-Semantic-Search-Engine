# Chunking Subtitles and Generating Embeddings

```python
import pandas as pd

df_subtitle=pd.read_csv('/content/drive/MyDrive/subtitle_data.csv')

df_subtitle

def custom_chunker(chunk_size, unique_id, content, name, chunk_buffer):
    # Split content into tokens
    tokens = [token for token in content.split()]
    # Get the length of the tokens
    num_tokens = len(tokens)

    # Initialize index and count
    start_index, chunk_count = 0, 0

    while True:
        if start_index == 0:
            chunk_start, chunk_end = start_index,chunk_buffer
            if chunk_end > num_tokens:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:]))
                names.append(name)  # Append the name here
                break
            else:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:chunk_end]))
                names.append(name)  # Append the name here
                start_index = chunk_end
        else:
            chunk_start, chunk_end = start_index - chunk_size, start_index + chunk_buffer - chunk_size
            if chunk_end > num_tokens:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:]))
                names.append(name)  # Append the name here
                break
            else:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:chunk_end]))
                names.append(name)  # Append the name here
                start_index = chunk_end
        chunk_count += 1

```

```python
new_id = []
new_content = []
names = []

for i in range(len(df_subtitle)):
    custom_chunker(25, i, df_subtitle['subtitle_processed'][i], df_subtitle['name'][i], 250)
```

```python
new_df = pd.DataFrame({'id': new_id, 'subtitle_content': new_content, 'name': names})
print(new_df)
```

```python
# Generating Sentence Embeddings
from sentence_transformers import SentenceTransformer


model =SentenceTransformer('all-MiniLM-L6-v2')
subtitles_convert=new_df["subtitle_content"].tolist()

subtitle_embds=model.encode(subtitles_convert)
```

# Initializing Chroma Database

```python

pip install chromadb
```

```python
import chromadb
from chromadb.utils import embedding_functions
```

```python
CHROMA_DATA_PATH = "D:\chromo database\store_embeddings"
client = chromadb.PersistentClient(path=CHROMA_DATA_PATH)
```

```python
COLLECTION_NAME = "subtitle_embeddings"

try:
    client.delete_collection(name=COLLECTION_NAME)
    print(f"Deleted existing collection '{COLLECTION_NAME}'")
except Exception as e:
    print(f"Error deleting collection '{COLLECTION_NAME}': {e}")

# Create a new collection
collection = client.create_collection(
    name=COLLECTION_NAME,
    metadata={"hnsw:space": "cosine"}
)

# Add documents to the new collection
for index, row in new_df.iterrows():
    embedding = subtitle_embds[index]  # Get the embedding for the current row
    document_name = row['name']
    document_id = row['id']

    # Add document to the collection
    collection.add(
        documents=[document_name],
        ids=[str(document_id)],
        embeddings=[embedding.tolist()]
    )
```
